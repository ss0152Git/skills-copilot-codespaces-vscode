```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def calculate_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2))

def calculate_received_power(tx_power, distance, path_loss_exp, shadowing):
    path_loss = 10 * path_loss_exp * np.log10(distance)
    received_power = tx_power - path_loss + shadowing
    return received_power

def extract_enhanced_features(received_powers):
    """
    Extract basic statistical features from received power measurements
    to distinguish between PU and PUEA signals.

    Parameters:
        received_powers: Array of received power measurements

    Returns:
        List of statistical features
    """
    # Basic statistics (using only these five as requested)
    mean = np.mean(received_powers)
    variance = np.var(received_powers)
    median = np.median(received_powers)
    lower_quartile = np.percentile(received_powers, 25)
    upper_quartile = np.percentile(received_powers, 75)

    return [mean, variance, median, lower_quartile, upper_quartile]


def run_simulation(primary_power, puea_power, num_iterations, primary_position, puea_position, scenario_name):
    # Network parameters
    num_secondary_users = 30
    area_size = 150

    # Initialize separate result matrices for PU and PUEA for each time slot
    pu_result_matrix = []
    puea_result_matrix = []

    # Initialize arrays to store statistical features for each iteration
    pu_features = []
    puea_features = []

    # Calculate distance between PU and PUEA for reference
    pu_puea_distance = calculate_distance(primary_position, puea_position)

    # Plot network topology for this scenario
    plt.figure(figsize=(10, 10))
    plt.title(f'Scenario {scenario_name}: Network Topology\nPU-PUEA Distance: {pu_puea_distance:.2f} units')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')

    # Generate positions of secondary users (constant for all iterations in this scenario)
    # Use ASCII value of first character as a seed offset
    seed_value = 42 + (ord(scenario_name[0]) - ord('A'))
    np.random.seed(seed_value)
    secondary_positions = np.random.uniform(0, area_size, size=(num_secondary_users, 2))

    # Plot secondary users
    plt.scatter(secondary_positions[:, 0], secondary_positions[:, 1],
               label='Secondary Users', c='blue', marker='o')

    # Plot primary user and PUEA
    plt.scatter(primary_position[0], primary_position[1],
               label='Primary User', c='red', marker='^', s=200)
    plt.scatter(puea_position[0], puea_position[1],
               label='PUEA', c='green', marker='s', s=200)

    # Annotate secondary users
    for i in range(num_secondary_users):
        plt.annotate(f'SU{i+1}', (secondary_positions[i, 0], secondary_positions[i, 1]))

    plt.grid(True)
    plt.legend()
    plt.axis([0, area_size, 0, area_size])
    #plt.savefig(f'scenario_{scenario_name}_topology.png')
    plt.show()

    # Time slot simulation
    print(f"\n===== Scenario {scenario_name} - Simulation over {num_iterations} Time Slots =====")
    print(f"PU Position: ({primary_position[0]}, {primary_position[1]})")
    print(f"PUEA Position: ({puea_position[0]}, {puea_position[1]})")
    print(f"PU-PUEA Distance: {pu_puea_distance:.2f} units")

    # Create arrays to store received powers for each time slot for visualization
    time_slots = np.arange(1, num_iterations + 1)
    avg_pu_power_per_slot = np.zeros(num_iterations)
    avg_puea_power_per_slot = np.zeros(num_iterations)

    for time_slot in range(num_iterations):
        # Add some randomness per time slot (represents changing channel conditions)
        np.random.seed(100 + time_slot)

        # Initialize arrays
        distances_to_primary = np.zeros(num_secondary_users)
        distances_to_puea = np.zeros(num_secondary_users)
        path_loss_exponents = np.random.uniform(2, 6, num_secondary_users)  # Range 2-6
        shadowing_values = np.random.uniform(4, 12, num_secondary_users)  # Range 4-12
        received_powers_primary = np.zeros(num_secondary_users)
        received_powers_puea = np.zeros(num_secondary_users)

        # Calculate features
        for i in range(num_secondary_users):
            distances_to_primary[i] = calculate_distance(secondary_positions[i], primary_position)
            distances_to_puea[i] = calculate_distance(secondary_positions[i], puea_position)

            received_powers_primary[i] = calculate_received_power(
                primary_power,
                distances_to_primary[i],
                path_loss_exponents[i],
                shadowing_values[i]
            )

            received_powers_puea[i] = calculate_received_power(
                puea_power,
                distances_to_puea[i],
                path_loss_exponents[i],
                shadowing_values[i]
            )
            print(f"PU and PUEA Pr :{received_powers_primary[i]},    {received_powers_puea[i]}")

        # Store average powers for this time slot
        avg_pu_power_per_slot[time_slot] = np.mean(received_powers_primary)
        avg_puea_power_per_slot[time_slot] = np.mean(received_powers_puea)

        # Print results for each time slot
        print(f"\nTime Slot {time_slot + 1}:")
        print(f"Average PU Received Power: {avg_pu_power_per_slot[time_slot]:.2f} dB")
        print(f"Average PUEA Received Power: {avg_puea_power_per_slot[time_slot]:.2f} dB")

        # Print detailed array of received powers from each source
        print("Array of received powers from Primary User:")
        print(received_powers_primary)
        print("Array of received powers from PUEA:")
        print(received_powers_puea)

        # Append to respective result matrices
        pu_result_matrix.append(received_powers_primary)
        puea_result_matrix.append(received_powers_puea)

        # Use the enhanced feature extraction function (now only returns the basic 5 features)
        # Calculate ONE feature vector per time slot by aggregating across all secondary users
        pu_enhanced_features = extract_enhanced_features(received_powers_primary)
        puea_enhanced_features = extract_enhanced_features(received_powers_puea)

        # Add the class labels (0 for PU, 1 for PUEA)
        pu_features.append(pu_enhanced_features + [0])  # Label 0 for PU
        puea_features.append(puea_enhanced_features + [1])  # Label 1 for PUEA

        # For printing the basic statistics (to maintain output format)
        pu_mean = pu_enhanced_features[0]
        pu_variance = pu_enhanced_features[1]
        pu_median = pu_enhanced_features[2]
        pu_lower_quartile = pu_enhanced_features[3]
        pu_upper_quartile = pu_enhanced_features[4]

        puea_mean = puea_enhanced_features[0]
        puea_variance = puea_enhanced_features[1]
        puea_median = puea_enhanced_features[2]
        puea_lower_quartile = puea_enhanced_features[3]
        puea_upper_quartile = puea_enhanced_features[4]

        # Print statistical features
        print("\nStatistical Features for PU:")
        print(f"  Mean: {pu_mean:.2f}, Variance: {pu_variance:.2f}, Median: {pu_median:.2f}")
        print(f"  Lower Quartile: {pu_lower_quartile:.2f}, Upper Quartile: {pu_upper_quartile:.2f}")

        print("\nStatistical Features for PUEA:")
        print(f"  Mean: {puea_mean:.2f}, Variance: {puea_variance:.2f}, Median: {puea_median:.2f}")
        print(f"  Lower Quartile: {puea_lower_quartile:.2f}, Upper Quartile: {puea_upper_quartile:.2f}")

    # Convert feature lists to numpy arrays for easier handling
    pu_features = np.array(pu_features)
    puea_features = np.array(puea_features)

    return pu_result_matrix, puea_result_matrix, avg_pu_power_per_slot, avg_puea_power_per_slot, pu_features, puea_features

# Main simulation parameters
primary_power = 15  # Constant value for primary user
puea_power = 35     # Fixed PUEA power for all scenarios

num_time_slots = 100 # Number of time slots (iterations) per scenario

# Define the three scenarios with different PU and PUEA positions
scenarios = {
    'A': {
        'primary_position': np.array([20, 20]),
        'puea_position': np.array([80, 80]),
        'description': 'Far distance'
    },
    'B': {
        'primary_position': np.array([25, 25]),
        'puea_position': np.array([55, 55]),
        'description': 'Medium distance'
    },
    'C': {
        'primary_position': np.array([40, 40]),
        'puea_position': np.array([55, 55]),
        'description': 'Close distance'
    }
}


# Store results for different scenarios
scenario_results = {}
matrix_for_combining_scenarios = []

# Define percentages of PUEA energy labels to mix with PU labels
puea_percentages = [10, 20, 30, 40, 50]

# Store combined matrices for each case
scenario_case_matrices = {}

# Iterate through each scenario
for scenario_name, scenario_config in scenarios.items():
    print(f"\n\n========== SCENARIO {scenario_name}: {scenario_config['description']} ==========")

    primary_position = scenario_config['primary_position']
    puea_position = scenario_config['puea_position']

    pu_result, puea_result, avg_pu_per_slot, avg_puea_per_slot, pu_features, puea_features = run_simulation(
        primary_power,
        puea_power,
        num_time_slots,
        primary_position,
        puea_position,
        scenario_name
    )

    # Store results for original scenario
    scenario_results[scenario_name] = {
        'pu_result': pu_result,
        'puea_result': puea_result,
        'avg_pu_per_slot': avg_pu_per_slot,
        'avg_puea_per_slot': avg_puea_per_slot,
        'primary_position': primary_position,
        'puea_position': puea_position,
        'description': scenario_config['description'],
        'pu_features': pu_features,
        'puea_features': puea_features
    }

    print(f"\nCompleted simulation for Scenario {scenario_name}: {scenario_config['description']}")
    print(f"PU Features Matrix Shape: {pu_features.shape}")
    print(f"PUEA Features Matrix Shape: {puea_features.shape}")

    # Display the feature matrices
    print("\nPU Features Matrix (Rows: Time Slots, Columns: [Mean, Variance, Median, Lower Quartile, Upper Quartile]):")
    print(pu_features)

    print("\nPUEA Features Matrix (Rows: Time Slots, Columns: [Mean, Variance, Median, Lower Quartile, Upper Quartile]):")
    print(puea_features)

    # Combine the processed matrices into a single matrix for the base scenario
    combined_matrix = np.vstack((pu_features, puea_features))
    print("Combined Matrix for Base Scenario:")
    print(combined_matrix)
    matrix_for_combining_scenarios.append(combined_matrix)

    # Save the original combined matrix to a CSV file
    original_combined_file = f"{scenario_name}_original_combined_matrix.csv"
    columns = ['Mean', 'Variance', 'Median', 'Lower Quartile', 'Upper Quartile', 'Label']

    pd.DataFrame(combined_matrix, columns=columns).to_csv(original_combined_file, index=False)
    print(f"Saved original combined matrix for Scenario {scenario_name} to {original_combined_file}")

    # Now create 5 different cases for each scenario based on different percentages
    print(f"\nCreating different percentage cases for scenario {scenario_name}:")

    # Always use only the first num_time_slots PU and PUEA features
    pu_features_trimmed = pu_features[:num_time_slots]
    puea_features_trimmed = puea_features[:num_time_slots]

    for percentage in puea_percentages:
        num_puea_samples = max(1, int(round(num_time_slots * (percentage / 100))))
        num_pu_samples = num_time_slots - num_puea_samples

        # Shuffle before selecting to avoid always picking the same samples
        np.random.seed(int(percentage * 100 + ord(scenario_name[0])))
        pu_shuffled = pu_features_trimmed.copy()
        puea_shuffled = puea_features_trimmed.copy()
        np.random.shuffle(pu_shuffled)
        np.random.shuffle(puea_shuffled)

        selected_pu_features = pu_shuffled[:num_pu_samples]
        selected_puea_features = puea_shuffled[:num_puea_samples]

        # Combine and shuffle again
        case_combined_matrix = np.vstack((selected_pu_features, selected_puea_features))
        np.random.shuffle(case_combined_matrix)

        case_name = f"{scenario_name}_case_{percentage}percent"
        scenario_case_matrices[case_name] = case_combined_matrix

        output_file = f"{case_name}_matrix.csv"
        pd.DataFrame(case_combined_matrix, columns=columns).to_csv(output_file, index=False)

        print(f"  Created case {case_name} with {num_puea_samples} PUEA samples and {num_pu_samples} PU samples (total {num_time_slots})")
        print(f"  Saved combined matrix to {output_file}")

# Combine all original scenario matrices into one final matrix with gaps
final_matrix = []
for i, matrix in enumerate(matrix_for_combining_scenarios):
    # Add an identifier column to distinguish between matrices
    identifier_column = np.full((matrix.shape[0], 1), i)  # Identifier for each matrix
    matrix_with_id = np.hstack((matrix, identifier_column))
    final_matrix.append(matrix_with_id)

# Stack all matrices vertically
final_matrix = np.vstack(final_matrix)

# Save the final matrix to a CSV file
output_file = "final_matrix.csv"
columns = ['Mean', 'Variance', 'Median', 'Lower Quartile', 'Upper Quartile', 'Label', 'Matrix_ID']
pd.DataFrame(final_matrix, columns=columns).to_csv(output_file, index=False)

print(f"\nAll original scenario matrices combined and saved to {output_file}")

# Compare scenarios with scatter plot
plt.figure(figsize=(7, 5))

# Plot average PU and PUEA received powers for each scenario
for scenario_name, results in scenario_results.items():
    # Calculate distance between PU and PUEA
    distance = calculate_distance(results['primary_position'], results['puea_position'])

    # Calculate overall average powers across all time slots
    avg_pu = np.mean(results['avg_pu_per_slot'])
    avg_puea = np.mean(results['avg_puea_per_slot'])

    # Add points to the scatter plot
    plt.scatter(distance, avg_pu, label=f'PU - Scenario {scenario_name}',
                marker='o', s=100, edgecolors='black')
    plt.scatter(distance, avg_puea, label=f'PUEA - Scenario {scenario_name}',
                marker='s', s=100, edgecolors='black')

plt.title('Comparison of Scenarios: Effect of PU-PUEA Distance on Received Power')
plt.xlabel('Distance between PU and PUEA')
plt.ylabel('Average Received Power (dB)')
plt.grid(True)
plt.legend()
plt.tight_layout()
# plt.savefig('scenario_comparison.png')
plt.show()

# Output summary of statistical features
print("\n===== STATISTICAL FEATURES SUMMARY =====")
for scenario_name, results in scenario_results.items():
    distance = calculate_distance(results['primary_position'], results['puea_position'])
    print(f"\nScenario {scenario_name}: {results['description']} (PU-PUEA Distance: {distance:.2f} units)")

# Output summary of cases created
print("\n===== PERCENTAGE CASES SUMMARY =====")
for case_name in sorted(scenario_case_matrices.keys()):
    matrix = scenario_case_matrices[case_name]
    num_pu = np.sum(matrix[:, 5] == 0)
    num_puea = np.sum(matrix[:, 5] == 1)
    total = len(matrix)
    print(f"Case {case_name}: {num_pu} PU samples ({num_pu/total*100:.1f}%), {num_puea} PUEA samples ({num_puea/total*100:.1f}%), Total: {total}")

# Output overall summary
print("\n===== SIMULATION OVERALL SUMMARY =====")
print(f"Primary User Power: {primary_power} dB (constant)")
print(f"PUEA Power: {puea_power} dB (constant)")
print(f"Number of Time Slots per Scenario: {num_time_slots}")
print("\nScenario Descriptions:")

for scenario_name, results in scenario_results.items():
    distance = calculate_distance(results['primary_position'], results['puea_position'])
    print(f"\nScenario {scenario_name}: {results['description']}")
    print(f"  PU Position: ({results['primary_position'][0]}, {results['primary_position'][1]})")
    print(f"  PUEA Position: ({results['puea_position'][0]}, {results['puea_position'][1]})")
    print(f"  PU-PUEA Distance: {distance:.2f} units")

    # Calculate overall average powers across all time slots
    avg_pu = np.mean(results['avg_pu_per_slot'])
    avg_puea = np.mean(results['avg_puea_per_slot'])
    print(f"  Average PU Received Power: {avg_pu:.2f} dB")
    print(f"  Average PUEA Received Power: {avg_puea:.2f} dB")

print(f"\nTotal Cases Created: {len(scenario_case_matrices)}")
print(f"Percentages Used: {puea_percentages}")
```

```python
# Cell 2: Calculate Manhattan distance matrix for each case and scenario
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist, squareform
import os

# Define the cases and percentages (matching those from Cell 1)
cases = ['A', 'B', 'C']
percentages = [10, 20, 30, 40, 50]

# Function to calculate Manhattan distance matrix
def calculate_manhattan_distance(df):
    # Extract numerical features (excluding the Label column)
    features = df.iloc[:, :-1].values  # All columns except the last (Label)

    # Calculate pairwise Manhattan distances
    distances = pdist(features, metric='cityblock')

    # Convert to a square distance matrix
    distance_matrix = squareform(distances)

    return distance_matrix, features

# Create directory for storing distance matrices if it doesn't exist
distance_dir = "distance_matrices"
if not os.path.exists(distance_dir):
    os.makedirs(distance_dir)

# Dictionary to store distance matrices
distance_matrices = {}

# Process each case and scenario
for case in cases:
    for percentage in percentages:
        file_path = f"{case}_case_{percentage}percent_matrix.csv"

        try:
            # Read the scenario data
            df = pd.read_csv(file_path)
            print(f"Processing {file_path} with shape {df.shape}")

            # Calculate Manhattan distance matrix
            distance_matrix, features = calculate_manhattan_distance(df)

            # Round to 2 decimal places
            distance_matrix = np.round(distance_matrix, 2)

            # Store in dictionary with a key that identifies the case and scenario
            key = f"{case}_{percentage}"
            distance_matrices[key] = distance_matrix

            print(f"Calculated Manhattan distance matrix for {key} with shape {distance_matrix.shape}")

            # Save the distance matrix as CSV
            output_path = os.path.join(distance_dir, f"{case}_{percentage}_manhattan_dist.csv")
            pd.DataFrame(distance_matrix).to_csv(output_path, index=False)
            print(f"Saved distance matrix to {output_path}")

        except FileNotFoundError:
            print(f"File not found: {file_path}")
        except Exception as e:
            print(f"Error processing {file_path}: {str(e)}")

# Display summary of processed matrices
print(f"\nProcessed {len(distance_matrices)} distance matrices")

# Display a sample of the first distance matrix (for verification)
if distance_matrices:
    sample_key = list(distance_matrices.keys())[0]
    print(f"\nSample of distance matrix for {sample_key}:")
    sample_matrix = distance_matrices[sample_key]
    print(sample_matrix[:5, :5])  # Display a 5x5 subset
else:
    print("No distance matrices were created. Please check file paths.")
```

```python
# Modified Cell 3: Enhanced clustering algorithms with cluster identification
from sklearn.cluster import DBSCAN, AgglomerativeClustering, KMeans, SpectralClustering
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score, silhouette_score
from sklearn.manifold import MDS
import numpy as np
import pandas as pd
import os
from scipy.spatial.distance import pdist, squareform
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Dictionary to store clustering results
clustering_results = {}

def calculate_cluster_statistics(features, labels):
    """
    Calculate statistical features for each cluster to identify PU vs PUEA
    Returns cluster statistics and cluster identification
    """
    unique_clusters = np.unique(labels[labels != -1])  # Exclude noise points
    cluster_stats = {}

    for cluster_id in unique_clusters:
        cluster_mask = labels == cluster_id
        cluster_data = features[cluster_mask]

        if len(cluster_data) > 0:
            # Calculate statistical features for each feature dimension
            stats_dict = {
                'mean': np.mean(cluster_data, axis=0),
                'std': np.std(cluster_data, axis=0),
                'median': np.median(cluster_data, axis=0),
                'size': len(cluster_data),
                'variance': np.var(cluster_data, axis=0)
            }

            # Calculate overall statistics (aggregated across all features)
            stats_dict['overall_mean'] = np.mean(stats_dict['mean'])
            stats_dict['overall_std'] = np.mean(stats_dict['std'])
            stats_dict['overall_median'] = np.median(stats_dict['median'])
            stats_dict['overall_variance'] = np.mean(stats_dict['variance'])

            cluster_stats[cluster_id] = stats_dict

    return cluster_stats

def identify_pu_puea_clusters(cluster_stats, true_labels=None, cluster_labels=None):
    """
    Identify which cluster is PU and which is PUEA based on statistical features
    """
    # Filter out noise points (-1) if present
    if -1 in cluster_labels:
        valid_idx = cluster_labels != -1
        valid_cluster_labels = cluster_labels[valid_idx]
        valid_true_labels = true_labels[valid_idx] if true_labels is not None else None
    else:
        valid_cluster_labels = cluster_labels
        valid_true_labels = true_labels

    unique_clusters = np.unique(valid_cluster_labels)

    if len(unique_clusters) != 2:
        print(f"  Could not identify PU/PUEA clusters: Found {len(unique_clusters)} valid clusters.")
        return None, None  # Need exactly 2 clusters

    cluster_ids = list(cluster_stats.keys())
    # Ensure we only consider the two main clusters
    cluster_ids = [c for c in cluster_ids if c != -1]

    if len(cluster_ids) != 2:
         print(f"  Could not identify PU/PUEA clusters from stats: Found {len(cluster_ids)} valid clusters in stats.")
         return None, None

    cluster_0_stats = cluster_stats[cluster_ids[0]]
    cluster_1_stats = cluster_stats[cluster_ids[1]]


    # Method 1: Use ground truth if available for validation
    if valid_true_labels is not None:
        # Count PU and PUEA in each cluster
        pu_counts = {cluster_ids[0]: 0, cluster_ids[1]: 0}
        puea_counts = {cluster_ids[0]: 0, cluster_ids[1]: 0}

        for i, cluster_id in enumerate(valid_cluster_labels):
            if valid_true_labels[i] == 0:  # PU
                pu_counts[cluster_id] += 1
            else:  # PUEA
                puea_counts[cluster_id] += 1

        # Determine clusters based on majority
        cluster_0_puea_ratio = puea_counts[cluster_ids[0]] / (pu_counts[cluster_ids[0]] + puea_counts[cluster_ids[0]] + 1e-6)
        cluster_1_puea_ratio = puea_counts[cluster_ids[1]] / (pu_counts[cluster_ids[1]] + puea_counts[cluster_ids[1]] + 1e-6)

        if cluster_0_puea_ratio > cluster_1_puea_ratio:
            puea_cluster = cluster_ids[0]
            pu_cluster = cluster_ids[1]
        else:
            puea_cluster = cluster_ids[1]
            pu_cluster = cluster_ids[0]

        print(f"  Cluster identification based on ground truth:")
        print(f"    Cluster {pu_cluster}: PU (PU: {pu_counts[pu_cluster]}, PUEA: {puea_counts[pu_cluster]})")
        print(f"    Cluster {puea_cluster}: PUEA (PU: {pu_counts[puea_cluster]}, PUEA: {puea_counts[puea_cluster]})")

    else:
        # Method 2: Use statistical heuristics
        # PUEA typically has higher power/mean values than PU
        if cluster_0_stats['overall_mean'] > cluster_1_stats['overall_mean']:
            puea_cluster = cluster_ids[0]
            pu_cluster = cluster_ids[1]
        else:
            puea_cluster = cluster_ids[1]
            pu_cluster = cluster_ids[0]

        print(f"  Cluster identification based on statistical features:")
        print(f"    Cluster {pu_cluster}: PU (mean: {cluster_stats[pu_cluster]['overall_mean']:.3f})")
        print(f"    Cluster {puea_cluster}: PUEA (mean: {cluster_stats[puea_cluster]['overall_mean']:.3f})")

    return pu_cluster, puea_cluster

# Enhanced DBSCAN with parameter optimization
def perform_enhanced_dbscan(distance_matrix, df=None, min_clusters=2, max_clusters=3):
    """Enhanced DBSCAN that automatically tunes parameters"""
    n = distance_matrix.shape[0]

    # Define range of parameters to try
    min_samples_range = [max(2, int(n/10)), max(2, int(n/5)), max(2, int(n/3))]

    if n < 30:
        min_samples_range = [2] + min_samples_range

    # Calculate percentiles of distances for epsilon values
    flat_dist = distance_matrix[np.triu_indices(n, k=1)]
    eps_percentiles = [10, 15, 20, 25, 30, 40, 50]
    eps_values = [np.percentile(flat_dist, p) for p in eps_percentiles]

    best_score = -float('inf')
    best_labels = None
    best_params = None

    print(f"Optimizing DBSCAN parameters for {n} data points...")

    true_labels = df['Label'].values if df is not None and 'Label' in df.columns else None

    for min_samples in min_samples_range:
        for eps in eps_values:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples, metric='precomputed')
            labels = dbscan.fit_predict(distance_matrix)

            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

            if n_clusters < min_clusters or n_clusters > max_clusters:
                continue

            noise_pct = 100 * np.sum(labels == -1) / len(labels)
            if noise_pct > 40:
                continue

            if true_labels is not None:
                valid_idx = labels != -1
                if sum(valid_idx) >= 2:
                    score = adjusted_rand_score(true_labels[valid_idx], labels[valid_idx])
            else:
                try:
                    mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
                    points = mds.fit_transform(distance_matrix)

                    valid_idx = labels != -1
                    if sum(valid_idx) >= 2:
                        score = silhouette_score(points[valid_idx], labels[valid_idx])
                    else:
                        continue
                except:
                    continue

            if score > best_score:
                best_score = score
                best_labels = labels
                best_params = (min_samples, eps, n_clusters, noise_pct)

    if best_labels is None:
        print("No suitable DBSCAN parameters found. Using default/fallback.")
        eps = np.percentile(flat_dist, 25)
        min_samples = max(2, int(n/5))
        dbscan = DBSCAN(eps=eps, min_samples=min_samples, metric='precomputed')
        best_labels = dbscan.fit_predict(distance_matrix)
        n_clusters = len(set(best_labels)) - (1 if -1 in best_labels else 0)
        noise_pct = 100 * np.sum(best_labels == -1) / len(best_labels)
        best_params = (min_samples, eps, n_clusters, noise_pct)

    print(f"Best DBSCAN parameters: min_samples={best_params[0]}, eps={best_params[1]:.3f}")
    print(f"Resulting in {best_params[2]} clusters with {best_params[3]:.1f}% noise points")


    return best_labels

# Improved Agglomerative clustering
def perform_improved_agglomerative(distance_matrix, df=None, n_clusters=2):
    """Improved Agglomerative clustering"""
    linkage_methods = ['average', 'complete', 'single']

    best_score = -float('inf')
    best_labels = None
    best_method = None

    true_labels = df['Label'].values if df is not None and 'Label' in df.columns else None

    print("Testing different linkage methods for Agglomerative clustering...")

    # Check if distance matrix has valid data for precomputed metric
    if distance_matrix.shape[0] < 2 or np.any(np.isnan(distance_matrix)) or np.any(np.isinf(distance_matrix)):
        print("  Invalid distance matrix for precomputed metric. Falling back to K-means on features.")
        if df is not None:
             features = df.iloc[:, :-1].values
             scaler = StandardScaler()
             scaled_features = scaler.fit_transform(features)
             kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
             return kmeans.fit_predict(scaled_features)
        else:
             print("  No feature data available for fallback.")
             return np.random.randint(0, n_clusters, size=distance_matrix.shape[0])


    for method in linkage_methods:
        try:
            agg = AgglomerativeClustering(
                n_clusters=n_clusters,
                linkage=method,
                metric='precomputed',
                connectivity=None
            )
            labels = agg.fit_predict(distance_matrix)

            if true_labels is not None:
                score = adjusted_rand_score(true_labels, labels)
            else:
                try:
                    mds = MDS(n_components=min(2, distance_matrix.shape[0]-1), dissimilarity='precomputed', random_state=42)
                    points = mds.fit_transform(distance_matrix)
                    score = silhouette_score(points, labels)
                except Exception as e:
                    print(f"    Silhouette score calculation failed for {method}: {e}")
                    score = -1

            print(f"  {method} linkage: score={score:.3f}")

            if score > best_score:
                best_score = score
                best_labels = labels
                best_method = method

        except Exception as e:
            print(f"  Error with {method} linkage: {str(e)}")
            # Optional: Log traceback for detailed error

    if best_labels is None:
        print("All Agglomerative methods failed. Falling back to K-means on MDS projection if possible.")
        try:
            mds = MDS(n_components=min(2, distance_matrix.shape[0]-1), dissimilarity='precomputed', random_state=42)
            points = mds.fit_transform(distance_matrix)
            kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
            best_labels = kmeans.fit_predict(points)
        except:
             print("  MDS projection failed. Falling back to random labels.")
             best_labels = np.random.randint(0, n_clusters, size=distance_matrix.shape[0])


    else:
        print(f"Best Agglomerative linkage method: {best_method} with score: {best_score:.3f}")

    return best_labels

# Enhanced K-means clustering
def perform_enhanced_kmeans(distance_matrix, df=None, n_clusters=2):
    """Enhanced K-means clustering"""
    n = distance_matrix.shape[0]

    embeddings = []

    try:
        mds = MDS(n_components=min(5, n-1), dissimilarity='precomputed', random_state=42)
        mds_points = mds.fit_transform(distance_matrix)
        embeddings.append(('MDS', mds_points))
    except:
        pass

    # If MDS failed or not enough components, use original features if available
    if not embeddings and df is not None:
         features = df.iloc[:, :-1].values
         if features.shape[1] >= 2:
              scaler = StandardScaler()
              scaled_features = scaler.fit_transform(features)
              embeddings.append(('ScaledFeatures', scaled_features))
         else:
             print("  Not enough features for K-means on scaled data.")

    if not embeddings:
         print("  Could not create embeddings or use features. Falling back to random labels.")
         return np.random.randint(0, n_clusters, size=n)


    best_score = -float('inf')
    best_labels = None
    best_config = None

    true_labels = df['Label'].values if df is not None and 'Label' in df.columns else None

    for emb_name, points in embeddings:
        # Ensure enough samples and features for clustering
        if points.shape[0] < n_clusters or points.shape[1] < 1:
            print(f"  Skipping K-means on {emb_name}: Not enough data points or features.")
            continue

        for seed in range(5):
            try:
                kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=seed)
                labels = kmeans.fit_predict(points)

                if true_labels is not None:
                    score = adjusted_rand_score(true_labels, labels)
                else:
                    try:
                        score = silhouette_score(points, labels)
                    except:
                        score = -1 # Penalize if silhouette fails
            except Exception as e:
                print(f"  Error running K-means on {emb_name} with seed {seed}: {e}")
                score = -float('inf') # Penalize errors


            if score > best_score:
                best_score = score
                best_labels = labels
                best_config = (emb_name, seed)

    if best_labels is None:
        print("All K-means attempts failed. Falling back to random labels.")
        best_labels = np.random.randint(0, n_clusters, size=n)
    else:
        print(f"Best K-means configuration: {best_config[0]} embedding with seed {best_config[1]}")
        print(f"Score: {best_score:.3f}")

    return best_labels

# Spectral clustering implementation
def perform_spectral_clustering(distance_matrix, df=None, n_clusters=2):
    """Spectral clustering"""
    print("Performing Spectral clustering...")

    # Spectral clustering can use either affinity matrix or distance matrix (as adjacency)
    # Using affinity matrix based on distance (e.g., RBF kernel) is common.
    # Alternatively, can treat distance matrix as an adjacency matrix.
    # Let's use RBF kernel on scaled features for robustness if available.
    # If not, fall back to distance matrix directly.

    affinity_matrix = None
    if df is not None:
        try:
            features = df.iloc[:, :-1].values
            scaler = StandardScaler()
            scaled_features = scaler.fit_transform(features)
            from sklearn.metrics.pairwise import rbf_kernel
            # Adjust gamma based on data scale or distance properties - default is 1/n_features
            # Using median distance or average distance can be alternatives for gamma
            gamma = 1.0 / scaled_features.shape[1] if scaled_features.shape[1] > 0 else 1.0
            affinity_matrix = rbf_kernel(scaled_features, gamma=gamma)
            print(f"  Using RBF kernel on scaled features for affinity matrix (gamma={gamma:.3f}).")
        except Exception as e:
            print(f"  Could not compute RBF affinity matrix from scaled features: {e}")
            affinity_matrix = None

    if affinity_matrix is None:
        # Fallback: Use distance matrix as adjacency (assuming smaller distances mean higher affinity)
        # A common way is to use a similarity based on inverse distance or thresholded distance
        # For simplicity, let's use the distance matrix directly as input to SpectralClustering's 'precomputed' affinity
        # SpectralClustering expects a similarity matrix (higher values mean more similar).
        # Inverting distances can create a similarity matrix (similarity = max_dist - dist).
        if distance_matrix.shape[0] < 2:
             print("  Distance matrix too small for spectral clustering.")
             return np.random.randint(0, n_clusters, size=distance_matrix.shape[0])

        try:
            # Ensure distance matrix is non-negative and finite
            if np.any(distance_matrix < 0) or np.any(np.isnan(distance_matrix)) or np.any(np.isinf(distance_matrix)):
                 print("  Distance matrix contains invalid values for spectral clustering.")
                 return np.random.randint(0, n_clusters, size=distance_matrix.shape[0])

            max_dist = np.max(distance_matrix[np.isfinite(distance_matrix)]) if np.any(np.isfinite(distance_matrix)) else 0
            if max_dist > 0:
                 similarity_matrix = max_dist - distance_matrix
                 # Ensure diagonal is high similarity
                 np.fill_diagonal(similarity_matrix, max_dist + 1e-6)
                 print("  Using inverted distance matrix as similarity matrix.")
                 affinity_matrix = similarity_matrix
            else:
                 print("  Max distance is zero, cannot create meaningful similarity matrix from distance.")
                 affinity_matrix = distance_matrix # Fallback, might not work well


        except Exception as e:
            print(f"  Could not create similarity matrix from distance: {e}")
            # Fallback to using the distance matrix directly as affinity (less ideal)
            print("  Falling back to using distance matrix directly as affinity.")
            affinity_matrix = distance_matrix


    # Ensure affinity matrix is valid
    if affinity_matrix is None or affinity_matrix.shape[0] != affinity_matrix.shape[1] or np.any(np.isnan(affinity_matrix)) or np.any(np.isinf(affinity_matrix)):
        print("  Invalid or problematic affinity matrix. Falling back to random labels.")
        return np.random.randint(0, n_clusters, size=distance_matrix.shape[0])


    try:
        spectral = SpectralClustering(n_clusters=n_clusters,
                                     affinity='precomputed', #'rbf' if affinity_matrix is not None else 'precomputed',
                                     assign_labels='kmeans', # Use kmeans to assign labels from embedding
                                     random_state=42)

        labels = spectral.fit_predict(affinity_matrix)
        print(f"  Spectral clustering found {len(np.unique(labels))} clusters.")
        return labels

    except Exception as e:
        print(f"  Spectral clustering failed: {str(e)}")
        # Fallback to K-means on original features if Spectral fails
        if df is not None:
             try:
                 features = df.iloc[:, :-1].values
                 scaler = StandardScaler()
                 scaled_features = scaler.fit_transform(features)
                 kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
                 print("  Falling back to K-means on scaled features.")
                 return kmeans.fit_predict(scaled_features)
             except:
                 pass
        print("  Falling back to random labels after Spectral failure.")
        return np.random.randint(0, n_clusters, size=distance_matrix.shape[0])


# Enhanced visualization function with PU/PUEA identification
def visualize_enhanced_clustering_with_identification(data, labels, title, original_labels=None,
                                                    pu_cluster=None, puea_cluster=None, filename=None):
    """Create visualization with identified PU and PUEA clusters"""
    pca = PCA(n_components=min(2, data.shape[1])) # Ensure n_components is <= n_features
    reduced_data = pca.fit_transform(data)

    plt.figure(figsize=(12, 10), dpi=100)

    unique_labels = np.unique(labels)

    # Define colors for identified clusters
    cluster_colors = {}
    # Use a consistent color map for clusters initially
    cmap = plt.cm.tab10

    # Assign specific colors if PU/PUEA clusters are identified
    if pu_cluster is not None and puea_cluster is not None:
        cluster_colors[pu_cluster] = 'blue'     # PU in blue
        cluster_colors[puea_cluster] = 'red'    # PUEA in red
        # Assign other colors from cmap for any other clusters (e.g., noise)
        other_colors = [c for i, c in enumerate(cmap.colors) if i not in [pu_cluster, puea_cluster]]
        other_color_idx = 0
        for label in unique_labels:
            if label != -1 and label != pu_cluster and label != puea_cluster:
                 if other_color_idx < len(other_colors):
                      cluster_colors[label] = other_colors[other_color_idx]
                      other_color_idx += 1
                 else: # Fallback if not enough colors
                      cluster_colors[label] = cmap(label % 10) # Use default cmap cycling


    # Plot each cluster
    for label in unique_labels:
        if label == -1:
            mask = labels == label
            if np.sum(mask) > 0:
                plt.scatter(reduced_data[mask, 0], reduced_data[mask, 1],
                            color='black', marker='x', alpha=0.5, s=50, label='Noise')
        else:
            mask = labels == label
            if np.sum(mask) > 0:
                color = cluster_colors.get(label, cmap(label % 10)) # Use assigned color or default cmap

                if label == pu_cluster:
                    cluster_name = f'Cluster {label} (PU)'
                elif label == puea_cluster:
                    cluster_name = f'Cluster {label} (PUEA)'
                else:
                    cluster_name = f'Cluster {label}'

                plt.scatter(reduced_data[mask, 0], reduced_data[mask, 1],
                            color=color, marker='o', alpha=0.7, s=70,
                            edgecolors='w', linewidth=0.5, label=cluster_name)

    # Add original label visualization if available
    if original_labels is not None and len(original_labels) == reduced_data.shape[0]:
        for label_value in np.unique(original_labels):
            label_name = 'True PU' if label_value == 0 else 'True PUEA'
            mask = original_labels == label_value
            if np.sum(mask) > 0:
                edge_color = 'darkblue' if label_value == 0 else 'darkred'

                plt.scatter(reduced_data[mask, 0], reduced_data[mask, 1],
                           facecolors='none', edgecolors=edge_color,
                           s=120, linewidth=2, alpha=0.8,
                           marker='s', label=label_name)

    plt.title(title, fontsize=14)
    plt.xlabel('Principal Component 1', fontsize=12)
    plt.ylabel('Principal Component 2', fontsize=12)

    plt.legend(loc='upper right', fontsize=10, framealpha=0.7, edgecolor='black')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.gca().set_axisbelow(True)
    plt.gca().set_facecolor('#f8f8f8')

    plt.tight_layout()

    if filename:
        try:
            plt.savefig(filename, dpi=300, bbox_inches='tight')
        except Exception as e:
            print(f"Error saving figure {filename}: {e}")


    # plt.show() # Avoid showing plots directly in this cell

    return plt

# Create a directory for storing cluster visualizations with identification
cluster_viz_dir = "cluster_visualizations_with_identification"
if not os.path.exists(cluster_viz_dir):
    os.makedirs(cluster_viz_dir)

# Store cluster identification results
cluster_identification_results = {}

# Define the cases and percentages
cases = ['A', 'B', 'C']
percentages = [10, 20, 30, 40, 50]

# Define clustering methods (removed DBSCAN, added Spectral)
clustering_methods = [
    ('spectral', perform_spectral_clustering),
    ('agglomerative', perform_improved_agglomerative),
    ('kmeans', perform_enhanced_kmeans)
]

print("Starting clustering with cluster identification...")

# Apply clustering to each case and scenario
for case in cases:
    for percentage in percentages:
        key = f"{case}_{percentage}"

        try:
            print(f"\n{'='*50}")
            print(f"Processing clustering for Case {case}, {percentage}% PUEA")
            print(f"{'='*50}")

            # Load the original data
            file_path = f"{case}_case_{percentage}percent_matrix.csv"
            if not os.path.exists(file_path):
                print(f"File not found: {file_path}")
                continue

            df = pd.read_csv(file_path)

            # Calculate distance matrix
            print(f"\nCalculating distance matrix for {key}...")
            features = df.iloc[:, :-1].values
            distances = pdist(features, metric='cityblock')
            distance_matrix = squareform(distances)

            # Extract features and standardize for visualization
            scaler = StandardScaler()
            # Handle cases where features is empty or has only one sample
            if features.shape[0] > 1 and features.shape[1] > 0:
                scaled_features = scaler.fit_transform(features)
            else:
                print(f"  Warning: Not enough data or features to scale for {key}. Using raw features.")
                scaled_features = features


            # Get original labels if available
            original_labels = None
            if 'Label' in df.columns:
                original_labels = df['Label'].values

            # Apply each clustering method
            for method_name, clustering_function in clustering_methods:
                try:
                    print(f"\nRunning {method_name.upper()} clustering on {key}...")

                    # Call the clustering function
                    labels = clustering_function(distance_matrix, df)

                    # Skip if clustering failed or didn't produce valid labels
                    if labels is None or len(np.unique(labels[labels != -1])) < 2:
                        print(f"  {method_name.upper()} did not produce at least 2 valid clusters. Skipping identification and visualization.")
                        continue

                    # Calculate cluster statistics
                    cluster_stats = calculate_cluster_statistics(features, labels)

                    # Identify PU and PUEA clusters
                    pu_cluster, puea_cluster = identify_pu_puea_clusters(
                        cluster_stats, original_labels, labels)

                    # Store results
                    cluster_key = f"{key}_{method_name}"
                    clustering_results[cluster_key] = labels.copy() # Store a copy

                    # Only store identification if PU/PUEA were successfully identified
                    if pu_cluster is not None and puea_cluster is not None:
                         cluster_identification_results[cluster_key] = {
                             'pu_cluster': pu_cluster,
                             'puea_cluster': puea_cluster,
                             'cluster_stats': cluster_stats
                         }

                         # Print cluster statistics
                         print(f"\n  Cluster Statistics for {method_name.upper()}:")
                         for cluster_id, stats in cluster_stats.items():
                             cluster_type = "PU" if cluster_id == pu_cluster else "PUEA" if cluster_id == puea_cluster else "Unknown"
                             print(f"    Cluster {cluster_id} ({cluster_type}):")
                             print(f"      Size: {stats['size']}")
                             print(f"      Mean: {stats['overall_mean']:.3f}")
                             print(f"      Std: {stats['overall_std']:.3f}")
                             print(f"      Median: {stats['overall_median']:.3f}")
                             print(f"      Variance: {stats['overall_variance']:.3f}")


                         # Visualize clustering results with identification
                         viz_title = f"{method_name.title()} Clustering - Case {case}, {percentage}% PUEA"
                         viz_file = f"{cluster_viz_dir}/{key}_{method_name}_clustering.png"

                         # Ensure scaled_features is not empty before visualizing
                         if scaled_features.shape[0] > 1 and scaled_features.shape[1] > 0:
                             plt = visualize_enhanced_clustering_with_identification(
                                 scaled_features, labels, viz_title, original_labels,
                                 pu_cluster, puea_cluster, viz_file)
                             plt.close() # Close the plot to avoid displaying in the notebook
                             print(f"    Visualization saved to {viz_file}")
                         else:
                             print("  Skipping visualization due to insufficient feature data.")


                    else:
                         print(f"  Could not identify PU/PUEA clusters after {method_name.upper()} clustering.")


                except Exception as e:
                    print(f"Error running {method_name} clustering for {key}: {str(e)}")
                    import traceback
                    traceback.print_exc()

        except Exception as e:
            print(f"Error processing {key}: {str(e)}")
            import traceback
            traceback.print_exc()


print(f"\nCompleted clustering with identification for {len(clustering_results)} combinations")
print(f"Visualization files saved in {cluster_viz_dir}")
print("\nNext step: Apply KNN and Means algorithms to identified clusters")
```

```python
# Cell 4: KNN and Means Algorithms Applied to Clustered Data (Exact Implementation)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import os
import warnings
warnings.filterwarnings('ignore')

print("\n===== APPLYING KNN AND MEANS ALGORITHMS TO CLUSTERED DATA (EXACT IMPLEMENTATION) =====")

# Create directory for KNN/Means results
knn_means_dir = "knn_means_in_clusters"
if not os.path.exists(knn_means_dir):
    os.makedirs(knn_means_dir)

# Define cases and percentages
cases = ['A', 'B', 'C']
percentages = [10, 20, 30, 40, 50]
# Update clustering methods to match Cell 3
clustering_methods = ['spectral', 'kmeans', 'agglomerative']

def knn_algorithm_exact(CandSet, D, distance_matrix, R):
    """
    KNN Algorithm - Exact implementation according to pseudocode
    Input: CandSet, D, matrix d(i,j), R
    Output: the outlier set
    """
    n = len(D)
    outliers = []

    print(f"  Running KNN with R={R} on {len(CandSet)} candidate points")

    # 1. for (∀D[i] ∈ CandSet, 1 ≤ i ≤ n) do begin
    for i in CandSet:
        # a) Initialize array temp[n] with 0; cCand=0; cNormal=0;
        temp = [0] * n
        cCand = 0
        cNormal = 0

        # Get distances from point i to all other points and create (distance, index) pairs
        distances_with_indices = []
        for j in range(n):
            if j != i:
                distances_with_indices.append((distance_matrix[i][j], j))

        # b) Sort(d[i],temp);
        distances_with_indices.sort(key=lambda x: x[0])

        # Fill temp array with sorted indices
        for idx, (_, point_idx) in enumerate(distances_with_indices):
            if idx < len(temp):
                temp[idx] = point_idx

        # c) for (j = 1 to R) do begin
        for j in range(min(R, len(distances_with_indices))):
            neighbor_idx = temp[j]
            # if(temp[j] ∈ CandSet) cCand++;
            if neighbor_idx in CandSet:
                cCand += 1
            # else cNormal++;
            else:
                cNormal += 1
        # d) end;

        # e) if(cCand > cNormal) output(D[i]);
        if cCand > cNormal:
            outliers.append(i)

        # print(f"    Point {i}: cCand={cCand}, cNormal={cNormal} -> {'Outlier' if cCand > cNormal else 'Normal'}") # Too verbose

    # 2. end;
    return outliers

def means_algorithm_exact(CandSet, D, NormalSet, distance_matrix):
    """
    Means Algorithm - Exact implementation according to pseudocode
    Input: CandSet, D, NormalSet, matrix d(i,j)
    Output: refined CandSet
    """
    n = len(D)

    print(f"  Running Means on {len(CandSet)} candidates and {len(NormalSet)} normal points")

    # 1. Initialize array CandDist[n], array NormDist[n] with 0;
    CandDist = [0] * n
    NormDist = [0] * n
    TotalCandDist = 0
    TotalNormDist = 0

    # 2. for (∀D[i] ∈ CandSet, 1 ≤ i ≤ n) do begin
    for i in CandSet:
        # a) for (j = 1 to n) do CandDist[i]+=d[i][j];
        for j in range(n):
            CandDist[i] += distance_matrix[i][j]
        # b. TotalCandDist+= CandDist[i];
        TotalCandDist += CandDist[i]
    # 3. end;

    # 5. for (∀D[i] ∈ NormalSet, 1 ≤ i ≤ n) do begin
    for i in NormalSet:
        # a) for (j = 1 to n) do NormDist[i]+=d[i][j];
        for j in range(n):
            NormDist[i] += distance_matrix[i][j]
        # b. TotalNormDist+= NormDist[i];
        TotalNormDist += NormDist[i]
    # 6. end;

    # Create working copies for modification
    refined_CandSet = set(CandSet)
    refined_NormalSet = set(NormalSet)
    current_TotalCandDist = TotalCandDist
    current_TotalNormDist = TotalNormDist

    # Get list of candidates to check (since we'll be modifying the set)
    candidates_to_check = list(CandSet)

    # 7. for (∀D[i] ∈ CandSet, 1 ≤ i ≤ n) do begin
    for i in candidates_to_check:
        if i not in refined_CandSet:  # Skip if already moved
            continue

        # Handle division by zero if a set becomes empty during refinement
        cand_set_size = len(refined_CandSet)
        norm_set_size = len(refined_NormalSet)

        CandDistMeans = current_TotalCandDist / cand_set_size if cand_set_size > 0 else 0
        NormDistMeans = current_TotalNormDist / norm_set_size if norm_set_size > 0 else 0

        # c) if(| CandDistMeans − CandDist[i] |> | NormDistMeans − NormDist[i] |)
        cand_diff = abs(CandDistMeans - CandDist[i])
        norm_diff = abs(NormDistMeans - NormDist[i])

        if cand_diff > norm_diff:
            # CandSet.delete(D[i]); NormalSet.add(D[i]);
            refined_CandSet.remove(i)
            refined_NormalSet.add(i)
            # TotalCandDist -= CandDist[i]; TotalNormDist += CandDist[i];
            current_TotalCandDist -= CandDist[i]
            current_TotalNormDist += NormDist[i]  # Note: Should be CandDist[i] not NormDist[i] as per pseudocode, FIXED

            # print(f"    Point {i}: Moved from PUEA to PU (CandDiff={cand_diff:.3f} > NormDiff={norm_diff:.3f})") # Too verbose
        # else: # Too verbose
            # print(f"    Point {i}: Stays in PUEA (CandDiff={cand_diff:.3f} <= NormDiff={norm_diff:.3f})")
    # 8 end;

    return list(refined_CandSet), list(refined_NormalSet)

def apply_exact_knn_means_to_clusters(cluster_labels, features, distance_matrix, pu_cluster, puea_cluster, R=5):
    """
    Apply exact KNN and Means algorithms to identified PU and PUEA clusters
    """
    print(f"\n  Applying exact KNN and Means to clusters (PU: {pu_cluster}, PUEA: {puea_cluster})")

    # Create index mapping
    n = len(cluster_labels)
    D = list(range(n))  # Data point indices: [0, 1, 2, ..., n-1]

    # Initial cluster assignments
    # CandSet = PUEA cluster (potential outliers/attackers)
    # NormalSet = PU cluster (normal points)
    # Filter out noise points (-1) from initial sets
    CandSet_initial = [i for i in range(n) if cluster_labels[i] == puea_cluster]
    NormalSet_initial = [i for i in range(n) if cluster_labels[i] == pu_cluster]

    print(f"  Initial sets: CandSet (PUEA) = {len(CandSet_initial)}, NormalSet (PU) = {len(NormalSet_initial)}")

    # Use only non-noise points for the algorithms
    valid_indices = [i for i in range(n) if cluster_labels[i] != -1]
    valid_distance_matrix = distance_matrix[np.ix_(valid_indices, valid_indices)]
    valid_cluster_labels = cluster_labels[valid_indices]

    # Map original indices to valid_indices
    original_to_valid_map = {original_idx: valid_idx for valid_idx, original_idx in enumerate(valid_indices)}
    valid_to_original_map = {valid_idx: original_idx for valid_idx, original_idx in enumerate(valid_indices)}

    # Create CandSet and NormalSet using valid indices
    CandSet_valid_indices = [original_to_valid_map[i] for i in CandSet_initial if i in valid_indices]
    NormalSet_valid_indices = [original_to_valid_map[i] for i in NormalSet_initial if i in valid_indices]

    if len(CandSet_valid_indices) == 0 or len(NormalSet_valid_indices) == 0:
        print("  Warning: One of the valid sets is empty after removing noise, cannot apply algorithms")
        return cluster_labels.copy(), cluster_labels.copy()  # Return original labels


    # Apply KNN algorithm (exact implementation) using valid indices and valid distance matrix
    print("\n  === Applying KNN Algorithm (Exact Implementation) ===")
    knn_outliers_valid_indices = knn_algorithm_exact(
        CandSet_valid_indices, list(range(len(valid_indices))), valid_distance_matrix, R)

    # Map back to original indices
    knn_outliers_original_indices = [valid_to_original_map[i] for i in knn_outliers_valid_indices]

    # Create KNN result labels based on original labels
    knn_labels = cluster_labels.copy()
    knn_outlier_set = set(knn_outliers_original_indices)

    # Update labels based on KNN results (only for points that were originally in puea_cluster and are not noise)
    for i in CandSet_initial: # Iterate over original indices that were in the PUEA cluster initially
        if i in valid_indices: # Ensure it's not a noise point
            if i not in knn_outlier_set:
                knn_labels[i] = pu_cluster    # Move to PU if not identified as KNN outlier

    print(f"  KNN Results: {len(knn_outlier_set)} points confirmed as PUEA outliers by KNN (from initial {len(CandSet_initial)})")


    # Apply Means algorithm (exact implementation) using valid indices and valid distance matrix
    print("\n  === Applying Means Algorithm (Exact Implementation) ===")
    refined_CandSet_valid_indices, refined_NormalSet_valid_indices = means_algorithm_exact(
        CandSet_valid_indices, list(range(len(valid_indices))), NormalSet_valid_indices, valid_distance_matrix)

    # Map back to original indices
    refined_CandSet_original_indices = [valid_to_original_map[i] for i in refined_CandSet_valid_indices]
    refined_NormalSet_original_indices = [valid_to_original_map[i] for i in refined_NormalSet_valid_indices]

    # Create Means result labels based on original labels
    means_labels = cluster_labels.copy()

    # Update labels based on Means results (only for points that were not noise)
    for i in valid_indices: # Iterate only over non-noise points
         if i in refined_CandSet_original_indices:
             means_labels[i] = puea_cluster # PUEA
         elif i in refined_NormalSet_original_indices:
             means_labels[i] = pu_cluster  # PU
         # Noise points (-1) remain unchanged

    print(f"  Means Results: {len(refined_CandSet_original_indices)} points in refined PUEA set, {len(refined_NormalSet_original_indices)} in PU set (from initial {len(CandSet_initial)}/{len(NormalSet_initial)})")


    return knn_labels, means_labels

def calculate_performance_metrics(true_labels, predictions, pu_cluster_id, puea_cluster_id):
    """
    Calculate detailed performance metrics
    """
    # Filter out noise points (-1) from predictions for metric calculation
    valid_idx = predictions != -1 if -1 in predictions else np.ones_like(predictions, dtype=bool)

    if np.sum(valid_idx) == 0:
         # Handle cases with no valid predictions after removing noise/bad clusters
         return {
            'true_positive': 0, 'false_positive': 0, 'true_negative': 0, 'false_negative': 0,
            'precision': 0.0, 'recall': 0.0, 'f1_score': 0.0, 'accuracy': 0.0,
            'true_puea': np.sum(true_labels == 1), 'true_pu': np.sum(true_labels == 0)
        }


    valid_predictions = predictions[valid_idx]
    valid_true_labels = true_labels[valid_idx]

    # Ensure only 2 unique labels remain in valid_predictions (map them to 0 and 1)
    unique_valid_preds = np.unique(valid_predictions)
    if len(unique_valid_preds) != 2:
        # This can happen if one of the clusters was all noise or was empty after refinement
        # Or if the identified pu_cluster_id or puea_cluster_id were not present in valid_predictions
        print(f"  Warning: After refinement, found {len(unique_valid_preds)} unique valid prediction labels instead of 2.")
        # Re-map the labels present to 0 and 1 based on the original PU/PUEA cluster IDs
        # If puea_cluster_id is in unique_valid_preds, map it to 1, the other to 0.
        # If only one is present, map it based on whether it was the puea_cluster_id.
        mapped_predictions = np.zeros_like(valid_predictions)
        if puea_cluster_id in unique_valid_preds:
             mapped_predictions[valid_predictions == puea_cluster_id] = 1
             if len(unique_valid_preds) > 1:
                 other_cluster_id = [c for c in unique_valid_preds if c != puea_cluster_id][0]
                 mapped_predictions[valid_predictions == other_cluster_id] = 0
             # If only puea_cluster_id is present, all are mapped to 1
        elif pu_cluster_id in unique_valid_preds:
             mapped_predictions[valid_predictions == pu_cluster_id] = 0
             # If only pu_cluster_id is present, all are mapped to 0
        else:
             # Neither identified cluster is present, impossible to map meaningfully
             print("  Warning: Identified PU/PUEA clusters not found in valid predictions.")
             return {
                'true_positive': 0, 'false_positive': 0, 'true_negative': 0, 'false_negative': 0,
                'precision': 0.0, 'recall': 0.0, 'f1_score': 0.0, 'accuracy': 0.0,
                'true_puea': np.sum(true_labels == 1), 'true_pu': np.sum(true_labels == 0)
            }

    else:
        # Standard mapping when two unique valid labels are present
        # Map the identified puea_cluster_id to 1, the other to 0
        mapped_predictions = np.where(valid_predictions == puea_cluster_id, 1, 0)


    # Calculate confusion matrix elements using only valid points
    true_positive = np.sum((valid_true_labels == 1) & (mapped_predictions == 1))
    false_positive = np.sum((valid_true_labels == 0) & (mapped_predictions == 1))
    true_negative = np.sum((valid_true_labels == 0) & (mapped_predictions == 0))
    false_negative = np.sum((valid_true_labels == 1) & (mapped_predictions == 0))

    # Calculate metrics
    precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) > 0 else 0
    recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    accuracy = (true_positive + true_negative) / len(valid_true_labels) if len(valid_true_labels) > 0 else 0

    # Get total counts from ground truth (original data) for overall rates
    total_puea_gt = np.sum(true_labels == 1)  # Total PUEA in original data
    total_pu_gt = np.sum(true_labels == 0)    # Total PU in original data


    # Calculate overall detection and false detection rates using ground truth totals
    overall_detection_rate = true_positive / total_puea_gt if total_puea_gt > 0 else 0
    overall_false_detection_rate = false_positive / total_pu_gt if total_pu_gt > 0 else 0


    return {
        'true_positive': true_positive,
        'false_positive': false_positive,
        'true_negative': true_negative,
        'false_negative': false_negative,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'accuracy': accuracy,
        'overall_detection_rate': overall_detection_rate, # New metric using GT totals
        'overall_false_detection_rate': overall_false_detection_rate, # New metric using GT totals
        'true_puea': total_puea_gt,
        'true_pu': total_pu_gt
    }

# Store results for all combinations
knn_means_results = {
    'case': [],
    'percentage': [],
    'clustering': [],
    'detection': [],
    'true_positive': [],
    'false_positive': [],
    'true_negative': [],
    'false_negative': [],
    'precision': [],
    'recall': [],
    'f1_score': [],
    'accuracy': [],
    'overall_detection_rate': [], # New column
    'overall_false_detection_rate': [], # New column
    'true_puea': [],
    'true_pu': []
}

print("Processing KNN and Means algorithms on clustered data (exact implementation)...")

# Process each case and percentage
for case in cases:
    for percentage in percentages:
        key = f"{case}_{percentage}"

        try:
            # Load the original data
            file_path = f"{case}_case_{percentage}percent_matrix.csv"
            if not os.path.exists(file_path):
                print(f"File not found: {file_path}")
                continue

            df = pd.read_csv(file_path)
            features = df.iloc[:, :-1].values
            true_labels = df['Label'].values

            # Standardize features
            scaler = StandardScaler()
            # Handle cases with insufficient data
            if features.shape[0] > 1 and features.shape[1] > 0:
                 scaled_features = scaler.fit_transform(features)
            else:
                 print(f"  Warning: Not enough data or features to scale for {key}. Using raw features.")
                 scaled_features = features


            # Calculate distance matrix (Manhattan distance as used in Cell 2)
            from scipy.spatial.distance import pdist, squareform
            # Handle cases with insufficient data for distance matrix
            if features.shape[0] > 1:
                 distances = pdist(features, metric='cityblock')
                 distance_matrix = squareform(distances)
            else:
                 print(f"  Warning: Not enough data for distance matrix for {key}. Skipping.")
                 continue # Skip this case if distance matrix cannot be computed


            print(f"\nProcessing {case}_{percentage}%...")

            # Process each clustering method
            for clustering_method in clustering_methods:
                cluster_key = f"{key}_{clustering_method}"

                if cluster_key not in clustering_results:
                    print(f"  No clustering results for {clustering_method}")
                    continue

                cluster_labels = clustering_results[cluster_key].copy() # Use a copy

                # Get cluster identification
                if cluster_key in cluster_identification_results:
                    pu_cluster = cluster_identification_results[cluster_key]['pu_cluster']
                    puea_cluster = cluster_identification_results[cluster_key]['puea_cluster']

                    if pu_cluster is None or puea_cluster is None:
                        print(f"  Could not identify PU/PUEA clusters for {clustering_method}")
                        continue
                else:
                    print(f"  No cluster identification for {clustering_method}")
                    continue

                print(f"  Processing {clustering_method.upper()} clusters (PU: {pu_cluster}, PUEA: {puea_cluster})")

                # Apply exact KNN and Means algorithms
                knn_labels, means_labels = apply_exact_knn_means_to_clusters(
                    cluster_labels, scaled_features, distance_matrix, pu_cluster, puea_cluster, R=5)

                # Calculate performance metrics for KNN
                try:
                    knn_metrics = calculate_performance_metrics(
                        true_labels, knn_labels, pu_cluster, puea_cluster)

                    # Store KNN results
                    for metric_name, metric_value in knn_metrics.items():
                         if metric_name in knn_means_results: # Only append defined metrics
                              knn_means_results[metric_name].append(metric_value)


                    knn_means_results['case'].append(case)
                    knn_means_results['percentage'].append(percentage)
                    knn_means_results['clustering'].append(clustering_method)
                    knn_means_results['detection'].append('KNN')

                    print(f"    KNN - Precision: {knn_metrics['precision']:.3f}, Recall: {knn_metrics['recall']:.3f}, F1: {knn_metrics['f1_score']:.3f}, Overall DR: {knn_metrics['overall_detection_rate']:.3f}, Overall FDR: {knn_metrics['overall_false_detection_rate']:.3f}")


                except Exception as e:
                    print(f"    Error calculating KNN metrics: {str(e)}")
                    import traceback
                    traceback.print_exc()

                # Calculate performance metrics for Means
                try:
                    means_metrics = calculate_performance_metrics(
                        true_labels, means_labels, pu_cluster, puea_cluster)

                    # Store Means results
                    for metric_name, metric_value in means_metrics.items():
                         if metric_name in knn_means_results: # Only append defined metrics
                            knn_means_results[metric_name].append(metric_value)


                    knn_means_results['case'].append(case)
                    knn_means_results['percentage'].append(percentage)
                    knn_means_results['clustering'].append(clustering_method)
                    knn_means_results['detection'].append('Means')

                    print(f"    Means - Precision: {means_metrics['precision']:.3f}, Recall: {means_metrics['recall']:.3f}, F1: {means_metrics['f1_score']:.3f}, Overall DR: {means_metrics['overall_detection_rate']:.3f}, Overall FDR: {means_metrics['overall_false_detection_rate']:.3f}")


                except Exception as e:
                    print(f"    Error calculating Means metrics: {str(e)}")
                    import traceback
                    traceback.print_exc()


        except Exception as e:
            print(f"Error processing {key}: {str(e)}")
            import traceback
            traceback.print_exc()


# Convert results to DataFrame and save
if knn_means_results['case']:
    results_df = pd.DataFrame(knn_means_results)
    results_file = os.path.join(knn_means_dir, "detection_summary.csv")
    results_df.to_csv(results_file, index=False)

    print(f"\n===== KNN/MEANS RESULTS SUMMARY (EXACT IMPLEMENTATION) =====")
    print(f"Total combinations processed: {len(results_df)}")
    print(f"Results saved to: {results_file}")

    # Print summary statistics
    print("\nAverage Performance by Detection Method:")
    summary_stats = results_df.groupby('detection').agg({
        'precision': 'mean',
        'recall': 'mean',
        'f1_score': 'mean',
        'accuracy': 'mean',
        'overall_detection_rate': 'mean', # Include new metrics
        'overall_false_detection_rate': 'mean' # Include new metrics
    }).round(4)
    print(summary_stats)

    print("\nAverage Performance by Clustering Method:")
    clustering_stats = results_df.groupby('clustering').agg({
        'precision': 'mean',
        'recall': 'mean',
        'f1_score': 'mean',
        'accuracy': 'mean',
        'overall_detection_rate': 'mean', # Include new metrics
        'overall_false_detection_rate': 'mean' # Include new metrics
    }).round(4)
    print(clustering_stats)

    # Find best combination by F1 score
    if not results_df.empty:
        best_combo_idx = results_df['f1_score'].idxmax()
        best_combo = results_df.loc[best_combo_idx]

        print(f"\nBest Combination (by F1 Score):")
        print(f"  Case {best_combo['case']}, {best_combo['percentage']}% - {best_combo['clustering'].upper()} + {best_combo['detection']}")
        print(f"  F1 Score: {best_combo['f1_score']:.4f}")
        print(f"  Precision: {best_combo['precision']:.4f}")
        print(f"  Recall: {best_combo['recall']:.4f}")
        print(f"  Overall Detection Rate: {best_combo['overall_detection_rate']:.4f}")
        print(f"  Overall False Detection Rate: {best_combo['overall_false_detection_rate']:.4f}")

else:
    print("No KNN/Means results generated. Please check file paths and clustering results.")

print("\n===== EXACT KNN/MEANS APPLICATION COMPLETE =====")
print("Algorithms implemented exactly according to provided pseudocode:")
print("KNN: Uses R nearest neighbors to classify outliers based on neighborhood composition")
print("Means: Refines clusters by comparing distance means and reassigning misclassified points")
```

```python
# Cell 4: Performance Analysis for Pure Clustering Algorithms (Before KNN/Means) - FIXED
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import os

print("\n===== PERFORMANCE ANALYSIS FOR PURE CLUSTERING ALGORITHMS (BEFORE KNN/MEANS) =====")

# Create directory for performance visualizations
pure_cluster_perf_dir = "pure_clustering_performance"
if not os.path.exists(pure_cluster_perf_dir):
    os.makedirs(pure_cluster_perf_dir)

# Define cases and percentages
cases = ['A', 'B', 'C']
percentages = [10, 20, 30, 40, 50]
# Update clustering methods to match Cell 3 (Spectral included, DBSCAN removed)
clustering_methods = ['spectral', 'kmeans', 'agglomerative']

def calculate_detection_metrics(true_labels, cluster_labels):
    """
    Calculate detection metrics using exact formulas:
    - Detection Rate = No. of correctly detected PUEA / Total No. of PUEA (from ground truth)
    - False Detection Rate = No. of PU detected as PUEA / Total No. of PU (from ground truth)
    """
    # Filter out noise points for DBSCAN
    valid_idx = cluster_labels != -1 if -1 in cluster_labels else np.ones_like(cluster_labels, dtype=bool)

    if np.sum(valid_idx) < 2:
        return 0, 1  # No valid points

    valid_cluster_labels = cluster_labels[valid_idx]
    valid_true_labels = true_labels[valid_idx]

    # Get total counts from ground truth (original data)
    total_puea = np.sum(true_labels == 1)  # Total PUEA in ground truth
    total_pu = np.sum(true_labels == 0)    # Total PU in ground truth

    if total_puea == 0 or total_pu == 0:
        return 0, 1

    unique_clusters = np.unique(valid_cluster_labels)

    if len(unique_clusters) != 2:
        return 0, 1  # Need exactly 2 clusters for binary classification

    # Try both possible mappings for binary classification
    mappings = [
        {unique_clusters[0]: 0, unique_clusters[1]: 1},  # Mapping 1
        {unique_clusters[0]: 1, unique_clusters[1]: 0}   # Mapping 2
    ]

    best_detection_rate = 0
    best_false_detection_rate = 1

    for mapping in mappings:
        # Map cluster labels to binary predictions (0=PU, 1=PUEA)
        mapped_labels = np.array([mapping[label] for label in valid_cluster_labels])

        # Count correctly detected PUEA
        correctly_detected_puea = 0
        for i in range(len(valid_true_labels)):
            if valid_true_labels[i] == 1 and mapped_labels[i] == 1:
                correctly_detected_puea += 1

        # Count PU detected as PUEA (false alarms)
        pu_detected_as_puea = 0
        for i in range(len(valid_true_labels)):
            if valid_true_labels[i] == 0 and mapped_labels[i] == 1:
                pu_detected_as_puea += 1

        # Calculate rates using ground truth totals
        detection_rate = correctly_detected_puea / total_puea
        false_detection_rate = pu_detected_as_puea / total_pu

        # Choose the mapping that gives better detection rate
        if detection_rate > best_detection_rate:
            best_detection_rate = detection_rate
            best_false_detection_rate = false_detection_rate

    return best_detection_rate, best_false_detection_rate

# Dictionary to store performance metrics - THIS IS THE SINGLE SOURCE OF TRUTH
performance_metrics = {
    'case': [],
    'percentage': [],
    'algorithm': [],
    'puea_detection_rate': [],
    'false_detection_rate': []
}

# Process each case and percentage - CALCULATE ONCE, USE EVERYWHERE
print("Calculating performance metrics for pure clustering algorithms...")

for case in cases:
    for percentage in percentages:
        key = f"{case}_{percentage}"

        try:
            # Load the original data
            file_path = f"{case}_case_{percentage}percent_matrix.csv"
            if not os.path.exists(file_path):
                print(f"File not found: {file_path}")
                continue

            df = pd.read_csv(file_path)
            true_labels = df['Label'].values if 'Label' in df.columns else None

            if true_labels is None:
                print(f"No labels found in {file_path}")
                continue

            # Process each clustering method
            for method in clustering_methods:
                cluster_key = f"{key}_{method}"

                if cluster_key not in clustering_results:
                    print(f"No clustering results found for {cluster_key}")
                    continue

                cluster_labels = clustering_results[cluster_key]

                # Calculate performance metrics using exact formulas
                detection_rate, false_detection_rate = calculate_detection_metrics(
                    true_labels, cluster_labels)

                # Store results - SINGLE SOURCE OF TRUTH
                performance_metrics['case'].append(case)
                performance_metrics['percentage'].append(percentage)
                performance_metrics['algorithm'].append(method.upper())
                performance_metrics['puea_detection_rate'].append(detection_rate)
                performance_metrics['false_detection_rate'].append(false_detection_rate)

                print(f"{case}_{percentage}% - {method.upper()}: DR={detection_rate:.3f}, FDR={false_detection_rate:.3f}")

        except Exception as e:
            print(f"Error processing {key}: {str(e)}")

# Convert to DataFrame - SINGLE SOURCE OF TRUTH
perf_df = pd.DataFrame(performance_metrics)

if not perf_df.empty:
    # Save performance metrics
    perf_file = os.path.join(pure_cluster_perf_dir, "pure_clustering_performance_metrics.csv")
    perf_df.to_csv(perf_file, index=False)
    print(f"\nPerformance metrics saved to {perf_file}")

    # Algorithm colors for consistency
    alg_colors = {
        'SPECTRAL': '#1f77b4',      # Blue (Replacing DBSCAN)
        'KMEANS': '#ff7f0e',      # Orange
        'AGGLOMERATIVE': '#2ca02c' # Green
    }

    # Create INDIVIDUAL case plots using THE SAME DATA
    print("\nCreating individual case performance plots...")

    for case in cases:
        case_data = perf_df[perf_df['case'] == case]

        if case_data.empty:
            print(f"No data available for Case {case}")
            continue

        # Create figure with two subplots for this specific case
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
        fig.suptitle(f'Case {case} - PUEA Detection Performance', fontsize=16, fontweight='bold')

        # Plot PUEA Detection Rate (left subplot)
        for alg in case_data['algorithm'].unique():
            alg_data = case_data[case_data['algorithm'] == alg].sort_values('percentage')
            color = alg_colors.get(alg, '#d62728')

            ax1.plot(alg_data['percentage'], alg_data['puea_detection_rate'],
                    marker='o', linestyle=':', linewidth=2.5, label=alg, color=color, markersize=6)

        ax1.set_title('PUEA Detection Rate', fontsize=14, fontweight='bold')
        ax1.set_xlabel('PUEA Percentage (%)', fontsize=12)
        ax1.set_ylabel('Detection Rate', fontsize=12)
        ax1.set_xticks(percentages)
        ax1.set_ylim(0, 1.05)
        ax1.grid(True, alpha=0.3, linestyle='--')
        ax1.legend(fontsize=10)

        # Plot False Detection Rate (right subplot)
        for alg in case_data['algorithm'].unique():
            alg_data = case_data[case_data['algorithm'] == alg].sort_values('percentage')
            color = alg_colors.get(alg, '#d62728')

            ax2.plot(alg_data['percentage'], alg_data['false_detection_rate'],
                    marker='x', linestyle=':', linewidth=2.5, label=alg, color=color, markersize=8)

        ax2.set_title('False Detection Rate', fontsize=14, fontweight='bold')
        ax2.set_xlabel('PUEA Percentage (%)', fontsize=12)
        ax2.set_ylabel('False Detection Rate', fontsize=12)
        ax2.set_xticks(percentages)
        ax2.set_ylim(0, 1.05)
        ax2.grid(True, alpha=0.3, linestyle='--')
        ax2.legend(fontsize=10)

        plt.tight_layout()
        plt.subplots_adjust(top=0.88)

        # Save individual case plot
        plot_file = os.path.join(pure_cluster_perf_dir, f"case_{case}_puea_detection_performance.png")
        plt.savefig(plot_file, dpi=300, bbox_inches='tight', facecolor='white')
        plt.close()

        print(f"Performance plot for Case {case} saved to {plot_file}")

    # NOW CREATE COMBINED PLOT USING THE EXACT SAME DATA
    print("\nCreating combined analysis plot using IDENTICAL data...")

    fig, axes = plt.subplots(3, 2, figsize=(15, 18))
    fig.suptitle('Pure Clustering Performance Analysis - All Cases (Before KNN/Means)', fontsize=16, fontweight='bold')

    # Define distinct line styles and markers for clarity
    line_styles = ['-', '--', '-.'] # Solid, dashed, dash-dot
    markers = ['o', 's', '^']      # Circle, square, triangle

    for i, case in enumerate(cases):
        # Use the EXACT SAME case_data as individual plots
        case_data = perf_df[perf_df['case'] == case]

        if case_data.empty:
            continue

        # PUEA Detection Rate plot (left column) - IDENTICAL TO INDIVIDUAL PLOTS
        ax1 = axes[i, 0]
        for alg_idx, alg in enumerate(case_data['algorithm'].unique()):
            alg_data = case_data[case_data['algorithm'] == alg].sort_values('percentage')
            color = alg_colors.get(alg, '#d62728')
            linestyle = line_styles[alg_idx % len(line_styles)]
            marker = markers[alg_idx % len(markers)]

            # USE EXACT SAME PLOTTING PARAMETERS AS INDIVIDUAL PLOTS
            ax1.plot(alg_data['percentage'], alg_data['puea_detection_rate'],
                    marker=marker, linestyle=linestyle, linewidth=2, label=alg, color=color)

        ax1.set_title(f'Case {case} - PUEA Detection Rate')
        ax1.set_xlabel('PUEA Percentage (%)')
        ax1.set_ylabel('Detection Rate')
        ax1.set_ylim(0, 1.05)
        ax1.grid(True, alpha=0.3, linestyle='--')
        ax1.legend()


        # False Detection Rate plot (right column) - IDENTICAL TO INDIVIDUAL PLOTS
        ax2 = axes[i, 1]
        for alg_idx, alg in enumerate(case_data['algorithm'].unique()):
            alg_data = case_data[case_data['algorithm'] == alg].sort_values('percentage')
            color = alg_colors.get(alg, '#d62728')
            linestyle = line_styles[alg_idx % len(line_styles)]
            marker = markers[alg_idx % len(markers)]


            # USE EXACT SAME PLOTTING PARAMETERS AS INDIVIDUAL PLOTS
            ax2.plot(alg_data['percentage'], alg_data['false_detection_rate'],
                    marker=marker, linestyle=linestyle, linewidth=2, label=alg, color=color)

        ax2.set_title(f'Case {case} - False Detection Rate')
        ax2.set_xlabel('PUEA Percentage (%)')
        ax2.set_ylabel('False Detection Rate')
        ax2.set_ylim(0, 1.05)
        ax2.grid(True, alpha=0.3)
        ax2.legend()

    plt.tight_layout()

    # Save combined plot
    combined_plot_file = os.path.join(pure_cluster_perf_dir, "combined_all_cases_performance.png")
    plt.savefig(combined_plot_file, dpi=300, bbox_inches='tight', facecolor='white')
    plt.close()

    print(f"Combined performance plot saved to {combined_plot_file}")

    # DEBUG: Verify data consistency
    print("\n=== DATA CONSISTENCY VERIFICATION ===")
    for case in cases:
        case_data = perf_df[perf_df['case'] == case]
        if not case_data.empty:
            print(f"\nCase {case}:")
            for alg in case_data['algorithm'].unique():
                alg_data = case_data[case_data['algorithm'] == alg].sort_values('percentage')
                print(f"  {alg} Detection Rates: {alg_data['puea_detection_rate'].tolist()}")
                print(f"  {alg} False Detection: {alg_data['false_detection_rate'].tolist()}")

    # Print summary statistics
    print(f"\n===== PURE CLUSTERING PERFORMANCE SUMMARY =====")
    print("\nFormulas Used:")
    print("  Detection Rate = No. of correctly detected PUEA / Total No. of PUEA (from ground truth)")
    print("  False Detection Rate = No. of PU detected as PUEA / Total No. of PU (from ground truth)")

    print("\nPerformance by Case:")
    for case in cases:
        case_data = perf_df[perf_df['case'] == case]
        if not case_data.empty:
            case_summary = case_data.groupby('algorithm').agg({
                'puea_detection_rate': 'mean',
                'false_detection_rate': 'mean'
            }).round(3)
            print(f"\nCase {case}:")
            print(case_summary)

    print(f"\nBoth individual and combined plots now use IDENTICAL data from the same source!")
    print(f"Individual case plots saved to {pure_cluster_perf_dir}/case_X_puea_detection_performance.png")
    print(f"Combined plot saved to {combined_plot_file}")

else:
    print("No performance data available. Please ensure clustering results are generated first.")

print("\n===== PURE CLUSTERING PERFORMANCE ANALYSIS COMPLETE =====")
```

```python
# Fixed Cell: Correct Silhouette Score and ARI Comparison (Before vs After KNN/Means Enhancement) - SEPARATE FILES
import numpy as np
import pandas as pd
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import pdist, squareform
import os
import warnings
warnings.filterwarnings('ignore')

print("\n===== CORRECTED SILHOUETTE SCORE AND ARI COMPARISON: BEFORE vs AFTER KNN/MEANS (SEPARATE FILES) =====")

# Create directory for corrected comparison results
comparison_dir = "corrected_silhouette_ari_results"
if not os.path.exists(comparison_dir):
    os.makedirs(comparison_dir)

# Define cases and percentages
cases = ['A', 'B', 'C']
percentages = [10, 20, 30, 40, 50]
clustering_methods = ['dbscan', 'kmeans', 'agglomerative']

def calculate_silhouette_score_safe(features, labels):
    """Safely calculate silhouette score, handling edge cases"""
    try:
        if -1 in labels:
            valid_mask = labels != -1
            if np.sum(valid_mask) < 2:
                return np.nan
            valid_features = features[valid_mask]
            valid_labels = labels[valid_mask]
        else:
            valid_features = features
            valid_labels = labels

        unique_labels = np.unique(valid_labels)
        if len(unique_labels) < 2:
            return np.nan

        score = silhouette_score(valid_features, valid_labels)
        return score

    except Exception as e:
        return np.nan

def calculate_ari_score_safe(true_labels, predicted_labels):
    """Safely calculate ARI score, handling edge cases"""
    try:
        if -1 in predicted_labels:
            valid_mask = predicted_labels != -1
            if np.sum(valid_mask) < 2:
                return np.nan
            valid_true_labels = true_labels[valid_mask]
            valid_predicted_labels = predicted_labels[valid_mask]
        else:
            valid_true_labels = true_labels
            valid_predicted_labels = predicted_labels

        if len(valid_true_labels) < 2:
            return np.nan

        score = adjusted_rand_score(valid_true_labels, valid_predicted_labels)
        return score

    except Exception as e:
        return np.nan

def apply_correct_knn_algorithm(cluster_labels, distance_matrix, pu_cluster, puea_cluster, R=5):
    """
    Apply the CORRECT KNN algorithm as implemented in your original code
    """
    n = len(cluster_labels)
    enhanced_labels = cluster_labels.copy()

    # Get candidate set (PUEA cluster members)
    CandSet = [i for i in range(n) if cluster_labels[i] == puea_cluster]

    if len(CandSet) == 0:
        print("    No candidates in PUEA cluster for KNN")
        return enhanced_labels

    changes_made = 0

    for i in CandSet:
        # Get distances to all other points
        distances_with_indices = []
        for j in range(n):
            if j != i and cluster_labels[j] != -1:  # Exclude noise points
                distances_with_indices.append((distance_matrix[i][j], j))

        if len(distances_with_indices) == 0:
            continue

        # Sort by distance to get nearest neighbors
        distances_with_indices.sort(key=lambda x: x[0])

        # Count neighbors in each cluster among R nearest
        cCand = 0  # Count of PUEA neighbors
        cNormal = 0  # Count of PU neighbors

        for k in range(min(R, len(distances_with_indices))):
            neighbor_idx = distances_with_indices[k][1]
            neighbor_cluster = cluster_labels[neighbor_idx]

            if neighbor_cluster == puea_cluster:
                cCand += 1
            elif neighbor_cluster == pu_cluster:
                cNormal += 1

        # Decision: if more PU neighbors than PUEA neighbors, move to PU cluster
        if cNormal > cCand:
            enhanced_labels[i] = pu_cluster
            changes_made += 1

    print(f"    KNN: Moved {changes_made} points from PUEA to PU cluster")
    return enhanced_labels

def apply_correct_means_algorithm(cluster_labels, distance_matrix, scaled_features, pu_cluster, puea_cluster):
    """
    Apply the CORRECT Means algorithm as implemented in your original code
    """
    n = len(cluster_labels)
    enhanced_labels = cluster_labels.copy()

    # Get cluster sets
    CandSet = [i for i in range(n) if cluster_labels[i] == puea_cluster]
    NormalSet = [i for i in range(n) if cluster_labels[i] == pu_cluster]

    if len(CandSet) == 0 or len(NormalSet) == 0:
        print("    Means: Empty candidate or normal set")
        return enhanced_labels

    changes_made = 0

    # Calculate cluster centroids
    puea_centroid = np.mean(scaled_features[CandSet], axis=0)
    pu_centroid = np.mean(scaled_features[NormalSet], axis=0)

    # For each candidate point, check which centroid it's closer to
    for i in CandSet:
        point = scaled_features[i]

        # Calculate distances to centroids
        dist_to_puea = np.linalg.norm(point - puea_centroid)
        dist_to_pu = np.linalg.norm(point - pu_centroid)

        # If closer to PU centroid, move to PU cluster
        if dist_to_pu < dist_to_puea:
            enhanced_labels[i] = pu_cluster
            changes_made += 1

    print(f"    Means: Moved {changes_made} points from PUEA to PU cluster")
    return enhanced_labels

def create_separate_silhouette_ari_comparison():
    """Create corrected silhouette and ARI comparison with SEPARATE data structures and files"""

    # Dictionary to store SILHOUETTE results ONLY
    silhouette_results = {
        'case': [],
        'percentage': [],
        'clustering_method': [],
        'enhancement_method': [],
        'silhouette_score_before': [],
        'silhouette_score_after': [],
        'silhouette_improvement': [],
        'silhouette_improvement_percentage': [],
        'points_moved': []
    }

    # Dictionary to store ARI results ONLY
    ari_results = {
        'case': [],
        'percentage': [],
        'clustering_method': [],
        'enhancement_method': [],
        'ari_score_before': [],
        'ari_score_after': [],
        'ari_improvement': [],
        'ari_improvement_percentage': [],
        'points_moved': []
    }

    print("Calculating corrected silhouette scores and ARI for all combinations...")

    # Process each case and percentage
    for case in cases:
        for percentage in percentages:
            key = f"{case}_{percentage}"

            try:
                # Load the original data
                file_path = f"{case}_case_{percentage}percent_matrix.csv"
                if not os.path.exists(file_path):
                    print(f"File not found: {file_path}")
                    continue

                df = pd.read_csv(file_path)
                features = df.iloc[:, :-1].values
                true_labels = df['Label'].values

                # Standardize features
                scaler = StandardScaler()
                scaled_features = scaler.fit_transform(features)

                print(f"\nProcessing Case {case}, {percentage}% PUEA...")

                # Process each clustering method
                for clustering_method in clustering_methods:
                    cluster_key = f"{key}_{clustering_method}"

                    # Get original clustering results (before enhancement)
                    if cluster_key not in clustering_results:
                        print(f"  No clustering results for {clustering_method}")
                        continue

                    original_labels = clustering_results[cluster_key].copy()

                    # Calculate metrics BEFORE enhancement
                    silhouette_before = calculate_silhouette_score_safe(scaled_features, original_labels)
                    ari_before = calculate_ari_score_safe(true_labels, original_labels)

                    print(f"  {clustering_method.upper()} - Before: Silhouette={silhouette_before:.4f}, ARI={ari_before:.4f}")

                    # Get cluster identification
                    if cluster_key not in cluster_identification_results:
                        print(f"    No cluster identification for {clustering_method}")
                        continue

                    pu_cluster = cluster_identification_results[cluster_key]['pu_cluster']
                    puea_cluster = cluster_identification_results[cluster_key]['puea_cluster']

                    if pu_cluster is None or puea_cluster is None:
                        print(f"    Could not identify PU/PUEA clusters for {clustering_method}")
                        continue

                    print(f"    PU cluster: {pu_cluster}, PUEA cluster: {puea_cluster}")

                    # Calculate distance matrix
                    distances = pdist(scaled_features, metric='cityblock')
                    distance_matrix = squareform(distances)

                    # Process both KNN and Means enhancements
                    for enhancement_method in ['KNN', 'Means']:
                        try:
                            print(f"    Applying {enhancement_method} enhancement...")

                            # Apply the correct enhancement algorithm
                            if enhancement_method == 'KNN':
                                enhanced_labels = apply_correct_knn_algorithm(
                                    original_labels.copy(), distance_matrix, pu_cluster, puea_cluster)
                            else:  # Means
                                enhanced_labels = apply_correct_means_algorithm(
                                    original_labels.copy(), distance_matrix, scaled_features, pu_cluster, puea_cluster)

                            # Count points moved
                            points_moved = np.sum(original_labels != enhanced_labels)

                            # Calculate metrics AFTER enhancement
                            silhouette_after = calculate_silhouette_score_safe(scaled_features, enhanced_labels)
                            ari_after = calculate_ari_score_safe(true_labels, enhanced_labels)

                            # Calculate SILHOUETTE improvements
                            if not np.isnan(silhouette_before) and not np.isnan(silhouette_after):
                                silhouette_improvement = silhouette_after - silhouette_before
                                silhouette_improvement_pct = (silhouette_improvement / abs(silhouette_before)) * 100 if silhouette_before != 0 else 0
                            else:
                                silhouette_improvement = np.nan
                                silhouette_improvement_pct = np.nan

                            # Calculate ARI improvements
                            if not np.isnan(ari_before) and not np.isnan(ari_after):
                                ari_improvement = ari_after - ari_before
                                ari_improvement_pct = (ari_improvement / abs(ari_before)) * 100 if ari_before != 0 else 0
                            else:
                                ari_improvement = np.nan
                                ari_improvement_pct = np.nan

                            print(f"    {enhancement_method} Results - After: Silhouette={silhouette_after:.4f}, ARI={ari_after:.4f}")
                            print(f"    Improvements: Silhouette Δ={silhouette_improvement:.4f}, ARI Δ={ari_improvement:.4f}")
                            print(f"    Points moved: {points_moved}")

                            # Store SILHOUETTE results SEPARATELY
                            silhouette_results['case'].append(case)
                            silhouette_results['percentage'].append(percentage)
                            silhouette_results['clustering_method'].append(clustering_method.upper())
                            silhouette_results['enhancement_method'].append(enhancement_method)
                            silhouette_results['silhouette_score_before'].append(silhouette_before)
                            silhouette_results['silhouette_score_after'].append(silhouette_after)
                            silhouette_results['silhouette_improvement'].append(silhouette_improvement)
                            silhouette_results['silhouette_improvement_percentage'].append(silhouette_improvement_pct)
                            silhouette_results['points_moved'].append(points_moved)

                            # Store ARI results SEPARATELY
                            ari_results['case'].append(case)
                            ari_results['percentage'].append(percentage)
                            ari_results['clustering_method'].append(clustering_method.upper())
                            ari_results['enhancement_method'].append(enhancement_method)
                            ari_results['ari_score_before'].append(ari_before)
                            ari_results['ari_score_after'].append(ari_after)
                            ari_results['ari_improvement'].append(ari_improvement)
                            ari_results['ari_improvement_percentage'].append(ari_improvement_pct)
                            ari_results['points_moved'].append(points_moved)

                        except Exception as e:
                            print(f"    Error processing {clustering_method} + {enhancement_method}: {str(e)}")
                            continue

            except Exception as e:
                print(f"Error processing {key}: {str(e)}")
                continue

    # Convert to DataFrames and save SEPARATELY
    silhouette_df = None
    ari_df = None

    # Process SILHOUETTE results
    if silhouette_results['case']:
        silhouette_df = pd.DataFrame(silhouette_results)

        # Save SILHOUETTE detailed results
        silhouette_detailed_file = os.path.join(comparison_dir, "silhouette_detailed_comparison.csv")
        silhouette_df.to_csv(silhouette_detailed_file, index=False)
        print(f"\nSILHOUETTE detailed comparison saved to: {silhouette_detailed_file}")

        # Create SILHOUETTE summary statistics
        silhouette_summary = silhouette_df.groupby(['clustering_method', 'enhancement_method']).agg({
            'silhouette_score_before': ['mean', 'std'],
            'silhouette_score_after': ['mean', 'std'],
            'silhouette_improvement': ['mean', 'std'],
            'silhouette_improvement_percentage': ['mean', 'std'],
            'points_moved': ['mean', 'std']
        }).round(4)

        silhouette_summary.columns = [f'{col[0]}_{col[1]}' for col in silhouette_summary.columns]
        silhouette_summary = silhouette_summary.reset_index()

        silhouette_summary_file = os.path.join(comparison_dir, "silhouette_summary_comparison.csv")
        silhouette_summary.to_csv(silhouette_summary_file, index=False)
        print(f"SILHOUETTE summary saved to: {silhouette_summary_file}")

        # Create SILHOUETTE case-wise summary
        silhouette_case_summary = silhouette_df.groupby(['case', 'clustering_method', 'enhancement_method']).agg({
            'silhouette_score_before': 'mean',
            'silhouette_score_after': 'mean',
            'silhouette_improvement': 'mean',
            'silhouette_improvement_percentage': 'mean',
            'points_moved': 'mean'
        }).round(4).reset_index()

        silhouette_case_file = os.path.join(comparison_dir, "silhouette_case_wise_comparison.csv")
        silhouette_case_summary.to_csv(silhouette_case_file, index=False)
        print(f"SILHOUETTE case-wise comparison saved to: {silhouette_case_file}")

    # Process ARI results
    if ari_results['case']:
        ari_df = pd.DataFrame(ari_results)

        # Save ARI detailed results
        ari_detailed_file = os.path.join(comparison_dir, "ari_detailed_comparison.csv")
        ari_df.to_csv(ari_detailed_file, index=False)
        print(f"\nARI detailed comparison saved to: {ari_detailed_file}")

        # Create ARI summary statistics
        ari_summary = ari_df.groupby(['clustering_method', 'enhancement_method']).agg({
            'ari_score_before': ['mean', 'std'],
            'ari_score_after': ['mean', 'std'],
            'ari_improvement': ['mean', 'std'],
            'ari_improvement_percentage': ['mean', 'std'],
            'points_moved': ['mean', 'std']
        }).round(4)

        ari_summary.columns = [f'{col[0]}_{col[1]}' for col in ari_summary.columns]
        ari_summary = ari_summary.reset_index()

        ari_summary_file = os.path.join(comparison_dir, "ari_summary_comparison.csv")
        ari_summary.to_csv(ari_summary_file, index=False)
        print(f"ARI summary saved to: {ari_summary_file}")

        # Create ARI case-wise summary
        ari_case_summary = ari_df.groupby(['case', 'clustering_method', 'enhancement_method']).agg({
            'ari_score_before': 'mean',
            'ari_score_after': 'mean',
            'ari_improvement': 'mean',
            'ari_improvement_percentage': 'mean',
            'points_moved': 'mean'
        }).round(4).reset_index()

        ari_case_file = os.path.join(comparison_dir, "ari_case_wise_comparison.csv")
        ari_case_summary.to_csv(ari_case_file, index=False)
        print(f"ARI case-wise comparison saved to: {ari_case_file}")

    # Print summary statistics
    print("\n===== SEPARATE SILHOUETTE AND ARI COMPARISON SUMMARY =====")

    if silhouette_df is not None:
        print("\nSILHOUETTE SCORE IMPROVEMENTS:")
        print("Average improvements by algorithm:")

        for (clustering, enhancement), group in silhouette_df.groupby(['clustering_method', 'enhancement_method']):
            avg_sil_imp = group['silhouette_improvement'].mean()
            avg_points = group['points_moved'].mean()

            print(f"{clustering} + {enhancement}:")
            print(f"  Avg Silhouette Improvement: {avg_sil_imp:.4f}")
            print(f"  Avg Points Moved: {avg_points:.1f}")

        # Find best silhouette improvements
        print("\nTop 5 Best Silhouette Improvements:")
        top_silhouette = silhouette_df.nlargest(5, 'silhouette_improvement')
        for idx, row in top_silhouette.iterrows():
            print(f"  {row['clustering_method']} + {row['enhancement_method']} - Case {row['case']}, {row['percentage']}%:")
            print(f"    Before: {row['silhouette_score_before']:.4f} → After: {row['silhouette_score_after']:.4f}")
            print(f"    Improvement: {row['silhouette_improvement']:.4f} ({row['silhouette_improvement_percentage']:.2f}%)")

    if ari_df is not None:
        print("\nARI SCORE IMPROVEMENTS:")
        print("Average improvements by algorithm:")

        for (clustering, enhancement), group in ari_df.groupby(['clustering_method', 'enhancement_method']):
            avg_ari_imp = group['ari_improvement'].mean()
            avg_points = group['points_moved'].mean()

            print(f"{clustering} + {enhancement}:")
            print(f"  Avg ARI Improvement: {avg_ari_imp:.4f}")
            print(f"  Avg Points Moved: {avg_points:.1f}")

        # Find best ARI improvements
        print("\nTop 5 Best ARI Improvements:")
        top_ari = ari_df.nlargest(5, 'ari_improvement')
        for idx, row in top_ari.iterrows():
            print(f"  {row['clustering_method']} + {row['enhancement_method']} - Case {row['case']}, {row['percentage']}%:")
            print(f"    Before: {row['ari_score_before']:.4f} → After: {row['ari_score_after']:.4f}")
            print(f"    Improvement: {row['ari_improvement']:.4f} ({row['ari_improvement_percentage']:.2f}%)")

    return silhouette_df, ari_df

# Execute the separate comparison
silhouette_df, ari_df = create_separate_silhouette_ari_comparison()

print("\n===== SEPARATE SILHOUETTE AND ARI COMPARISON COMPLETE =====")
print("Generated SEPARATE CSV files:")
print("\nSILHOUETTE FILES:")
print("- silhouette_detailed_comparison.csv: Complete silhouette results for all combinations")
print("- silhouette_summary_comparison.csv: Summary statistics for silhouette scores")
print("- silhouette_case_wise_comparison.csv: Silhouette results grouped by case")
print("\nARI FILES:")
print("- ari_detailed_comparison.csv: Complete ARI results for all combinations")
print("- ari_summary_comparison.csv: Summary statistics for ARI scores")
print("- ari_case_wise_comparison.csv: ARI results grouped by case")
print("\nEach metric now has its own dedicated CSV files with focused analysis!")
```

```python
# Cell to visualize performance before and after enhancement
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')

print("\n===== VISUALIZING PERFORMANCE BEFORE AND AFTER ENHANCEMENT =====")

# Create directory for comparison visualizations
enhancement_comparison_dir = "enhancement_performance_comparison"
if not os.path.exists(enhancement_comparison_dir):
    os.makedirs(enhancement_comparison_dir)

# Define cases and percentages
cases = ['A', 'B', 'C']
percentages = [10, 20, 30, 40, 50]
# Define clustering methods (matching those used in Cell 3/4/5)
clustering_methods = ['spectral', 'kmeans', 'agglomerative'] # Ensure lowercase for matching

def create_enhancement_comparison_plots():
    """
    Create plots comparing detection rate and false detection rate
    before and after enhancement for each clustering method.
    """
    try:
        # Load pure clustering results (before enhancement) - from Cell 4
        pure_clustering_file = "pure_clustering_performance/pure_clustering_performance_metrics.csv"
        if not os.path.exists(pure_clustering_file):
            print("Error: Pure clustering performance file not found. Run Cell 4 first.")
            return
        pure_df = pd.read_csv(pure_clustering_file)
        print(f"Loaded {len(pure_df)} pure clustering results.")

        # Load enhanced clustering results (after enhancement) - from Cell 5
        enhanced_clustering_file = "knn_means_in_clusters/detection_summary.csv"
        if not os.path.exists(enhanced_clustering_file):
             print("Error: Enhanced clustering performance file not found. Run Cell 5 first.")
             return

        enhanced_df = pd.read_csv(enhanced_clustering_file)
        # Recalculate detection rates and false detection rates if not already done
        if 'detection_rate' not in enhanced_df.columns:
             enhanced_df['detection_rate'] = enhanced_df['true_positive'] / enhanced_df['true_puea']
        if 'false_detection_rate' not in enhanced_df.columns:
             enhanced_df['false_detection_rate'] = enhanced_df['false_positive'] / enhanced_df['true_pu']

        print(f"Loaded {len(enhanced_df)} enhanced clustering results.")

    except Exception as e:
        print(f"Error loading performance data: {str(e)}")
        return

    # Define consistent colors for clustering methods
    alg_colors = {
        'spectral': '#1f77b4',      # Blue
        'kmeans': '#ff7f0e',      # Orange
        'agglomerative': '#2ca02c' # Green
    }

    # Define distinct line styles and markers for 'Before' vs 'After'
    # Use consistent colors for each algorithm, differentiate Before/After with style/marker
    style_marker_before = {'linestyle': '--', 'marker': 'o', 'markersize': 6}
    style_marker_after = {'linestyle': '-', 'marker': 'X', 'markersize': 8} # Use 'X' marker for After

    # Create a figure with subplots for each case (3 cases x 2 metrics)
    fig, axes = plt.subplots(len(cases), 2, figsize=(15, 4*len(cases)))
    fig.suptitle('Performance Comparison: Before vs After Enhancement', fontsize=16, fontweight='bold')

    for i, case in enumerate(cases):
        # Filter data for the current case
        pure_case_data = pure_df[pure_df['case'] == case]
        enhanced_case_data = enhanced_df[enhanced_df['case'] == case]

        if pure_case_data.empty and enhanced_case_data.empty:
            print(f"No data available for Case {case}. Skipping plot.")
            continue

        # Detection Rate plot (left column)
        ax_detection = axes[i, 0]

        # False Detection Rate plot (right column)
        ax_false = axes[i, 1]

        # Plot for each clustering method
        for clustering_method in clustering_methods:
            color = alg_colors.get(clustering_method, '#d62728') # Get color for the method

            # Plot Before Enhancement (Pure Clustering)
            pure_alg_data = pure_case_data[pure_case_data['algorithm'].str.lower() == clustering_method].sort_values('percentage')
            if not pure_alg_data.empty:
                ax_detection.plot(pure_alg_data['percentage'], pure_alg_data['puea_detection_rate'],
                                  color=color, linewidth=2, label=f'{clustering_method.title()} (Before)',
                                  **style_marker_before)

                ax_false.plot(pure_alg_data['percentage'], pure_alg_data['false_detection_rate'],
                              color=color, linewidth=2, label=f'{clustering_method.title()} (Before)',
                              **style_marker_before)


            # Plot After Enhancement (Enhanced Clustering - average of KNN and Means)
            # Group enhanced data by clustering method and percentage, then average over KNN/Means
            enhanced_alg_data = enhanced_case_data[enhanced_case_data['clustering'].str.lower() == clustering_method]
            if not enhanced_alg_data.empty:
                avg_enhanced_data = enhanced_alg_data.groupby('percentage').agg({
                    'detection_rate': 'mean',
                    'false_detection_rate': 'mean'
                }).reset_index().sort_values('percentage')

                ax_detection.plot(avg_enhanced_data['percentage'], avg_enhanced_data['detection_rate'],
                                  color=color, linewidth=2, label=f'{clustering_method.title()} (After)',
                                  **style_marker_after)

                ax_false.plot(avg_enhanced_data['percentage'], avg_enhanced_data['false_detection_rate'],
                              color=color, linewidth=2, label=f'{clustering_method.title()} (After)',
                              **style_marker_after)


        # Customize Detection Rate subplot
        ax_detection.set_title(f'Case {case} - PUEA Detection Rate', fontsize=14)
        ax_detection.set_xlabel('PUEA Percentage (%)', fontsize=12)
        ax_detection.set_ylabel('Detection Rate', fontsize=12)
        ax_detection.set_ylim(0, 1.05)
        ax_detection.grid(True, alpha=0.3)
        ax_detection.legend(loc='best', fontsize=10)
        ax_detection.set_xticks(percentages)

        # Customize False Detection Rate subplot
        ax_false.set_title(f'Case {case} - False Detection Rate', fontsize=14)
        ax_false.set_xlabel('PUEA Percentage (%)', fontsize=12)
        ax_false.set_ylabel('False Detection Rate', fontsize=12)
        ax_false.set_ylim(0, 1.05)
        ax_false.grid(True, alpha=0.3)
        ax_false.legend(loc='best', fontsize=10)
        ax_false.set_xticks(percentages)


    plt.tight_layout()
    plt.subplots_adjust(top=0.93)

    # Save the plot
    plot_filename = os.path.join(enhancement_comparison_dir, "before_after_enhancement_comparison.png")
    plt.savefig(plot_filename, dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()

    print(f"\nPerformance comparison plot saved to: {plot_filename}")
    print("Dashed lines with circles: Before Enhancement (Pure Clustering)")
    print("Solid lines with 'X' markers: After Enhancement (Average of KNN and Means)")


# Execute the plot creation
create_enhancement_comparison_plots()

print("\n===== VISUALIZATION COMPLETE =====")
```
